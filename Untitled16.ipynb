###1. 라이브러리 및 데이터

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

train_data = pd.read_csv('/content/drive/MyDrive/229611_버스 운행시간 예측 경진대회_data/train.csv')
test_data = pd.read_csv('/content/drive/MyDrive/229611_버스 운행시간 예측 경진대회_data/test.csv')
train_data.head(5)
```

###2. 데이터 전처리

**이상치 제거**

700이라는 수치는 가장 최고의 distance의 next_arrive_time이 600대였기 때문에 그 이외 너무 큰 시간을 소비하는 구간은 측정 오류로 판단하였다.

```python
train_data = train_data[train_data['next_arrive_time'] <= 700]
train_data.head()
#&lt; stands for lesser than (<) symbol
```

**정류장명 인코딩 작업**

XGBoost는 숫자 데이터만 취급하므로 정류장 이름을 LabelEncoder를 이용하여 치환시킨다.

```python
station_encoder = LabelEncoder()

_station = list(train_data['now_station'].values) + list(train_data['next_station'].values) #train_data의 모든 정류장 이름
station_set = set(_station) #set(): 집합 관련 함수.고유한값 가진다.
print(len(station_set))
station_encoder.fit(list(station_set))  #인코딩

#모든 학습, 시험 데이터의 정류장 정보 치환
train_data['now_station'] = station_encoder.transform(train_data['now_station'])
train_data['next_station'] = station_encoder.transform(train_data['next_station'])
test_data['now_station'] = station_encoder.transform(test_data['now_station'])
test_data['next_station'] = station_encoder.transform(test_data['next_station'])
train_data.head()
```

* labelencoding할 때 fit과 transform을 다른 데이터로.

### 4. 변수 선택 및 모델 구축

feature engineering & Initial Modeling

**날짜->요일->평일/주말 변환**

* date값을 이용할 수 있는 의미있는 값으로 변환
* datetime을 이용하여 각 날짜를 요일로 변환하고, 이를 바탕으로 평일과 주말 두가지로 분류.

```python
## train data
train_data['date'] = pd.to_datetime(train_data['date']) #date값을 datetime으로 변환
train_data['weekday'] = train_data['date'].dt.weekday #Monday 0, Sunday 6
train_data['weekday'] = train_data['weekday'].apply(lambda x: 0 if x < 5 else 1)
# 0~5는 월요일~금요일이므로 평일이면 0, 주말이면 1을 설정함.

train_data = pd.get_dummies(train_data, columns=['weekday']) #평일/주말에 대해 One-hot Encoding

train_data = train_data.drop('date',axis=1) #필요없는 date 칼럼을 drop
train_data.head()
```

```python
## test data
test_data['date'] = pd.to_datetime(test_data['date']) #date값을 datetime으로 변환
test_data['weekday'] = test_data['date'].dt.weekday #Monday 0, Sunday 6
test_data['weekday'] = test_data['weekday'].apply(lambda x: 0 if x < 5 else 1)
# 0~5는 월요일~금요일이므로 평일이면 0, 주말이면 1을 설정함.

test_data = pd.get_dummies(test_data, columns=['weekday']) #평일/주말에 대해 One-hot Encoding

test_data = test_data.drop('date',axis=1) #필요없는 date 칼럼을 drop
test_data.head()
```

**정류장 출발 시간 3분류 변환**

기존 시간을 24개로 두게 되면 모델의 복잡도가 증가할 수 있으므로 과적합을 막기 위해 아침/오후/저녁 3가지 분류로 변환한 후 One-hot Encoding.

```python
train_data['now_arrive_time'] <='05시'
```

```python
## train data
train_data['time_group']='group' #time_group 변수 미리 생성

train_data.loc[(train_data['now_arrive_time'] >='05시') & (train_data['now_arrive_time']<'12시'),['time_group']]='morning' #05~11시
train_data.loc[(train_data['now_arrive_time'] >='12시') & (train_data['now_arrive_time']<'18시'),['time_group']]='afternoon' #12~17시
train_data.loc[(train_data['now_arrive_time'] >='18시') | (train_data['now_arrive_time']=='00시'),['time_group']]='evening' #18~00시

train_data = pd.get_dummies(train_data,columns=['time_group']) # 원 핫 인코딩을 수행
train_data = train_data.drop('now_arrive_time',axis=1) #필요없는 now_arrive_time drop
train_data.head()
```

```python
## test data
test_data['time_group']='group' #time_group 변수 미리 생성

test_data.loc[(test_data['now_arrive_time'] >='05시') & (test_data['now_arrive_time']<'12시'),['time_group']]='morning' #05~11시
test_data.loc[(test_data['now_arrive_time'] >='12시') & (test_data['now_arrive_time']<'18시'),['time_group']]='afternoon' #12~17시
test_data.loc[(test_data['now_arrive_time'] >='18시') | (test_data['now_arrive_time']=='00시'),['time_group']]='evening' #18~00시

test_data = pd.get_dummies(test_data,columns=['time_group']) # 원 핫 인코딩을 수행
test_data = test_data.drop('now_arrive_time',axis=1) #필요없는 now_arrive_time drop
test_data.head()
```

**필요없는 관측치 제거**

* route_nm은 route_id와 같은 역할의 데이터이므로 중복으로 판단하여 제거. id는 학습에서 의미없는 데이터이므로 제거.
* latitude, longitude는 큰 의미가 없다고 판단하고, 모델의 복잡도만 높인다고 판단하여 제거.

```python
## train data
train_data = train_data.drop(['id', 'route_nm', 'next_latitude', 'next_longitude', 'now_latitude', 'now_longitude'],axis=1)
train_data.head()
```

```python
# test는 제출 파일 생성 시 id가 필요하니까 남겨둔다.
## test data

test_data = test_data.drop(['route_nm', 'next_latitude', 'next_longitude', 'now_latitude', 'now_longitude'],axis=1)
test_data.head()
```

**학습 데이터를 구성**

```python
# 학습 데이터 칼럼에서 목표치인 next_arrive_time만 제거하여 선택.
input_var = list(train_data.columns)
input_var.remove('next_arrive_time')

Xtrain = train_data[input_var] #학습 데이터 선택
Ytrain = train_data['next_arrive_time'] # target 값인 Y 데이터 선택
#Xtrain = train_data.drop(['next_arrive_time'],axis=1) 하면 더 간단하게 가능.

X_test = test_data[input_var] # test 데이터 선택
```

**모델 구축 및 학습 준비**

* 빠른 학습과 정밀도를 보장하는 XGBoost(eXtreme Gradient Boosting) 사용
* K-Fold를 이용하여 K-Validation 수행.

```python
model = XGBRegressor(raindom_state=110, verbosity=0, nthread=23, n_estimators=980,max_depth=4)
kfold = KFold(n_splits=8, shuffle=True, random_state=777)
n_iter=0
cv_score=[]

def rmse(target,pred):
  return np.sqrt(np.sum(np.power(target - pred,2)) / np.size(pred))
```

### 5. 모델 학습 및 검증

**모델 학습 & 검증 진행**

```python
for train_index, test_index in kfold.split(Xtrain, Ytrain):
    # K Fold가 적용된 train, test 데이터를 불러온다
    X_train, X_test = Xtrain.iloc[train_index,:], Xtrain.iloc[test_index, :]
    Y_train, Y_test = Ytrain.iloc[train_index], Ytrain.iloc[test_index]
    
    # 모델 학습과 예측 수행
    model.fit(X_train, Y_train)
    pred = model.predict(X_test)
    print(pred)
    
    # 정확도 RMSE 계산
    n_iter += 1
    score = rmse(Y_test, pred)
    print(score)
    cv_score.append(score)
print('\n교차 검증별 RMSE :', np.round(cv_score, 4))
print('평균 검증 RMSE :', np.mean(cv_score))
```

* GridSearchCV 이용(GridSearch + Cross Validation)
* KFold보다도 cross_val_score 이용이 더 간단하기도 하다(stratified K-fold 검증방법 사용)
