## ■ 넘파이 ndarray 개요

---

```python
import numpy as np
```

-ndarray: 넘파이의 기반 데이터 타입.

-ndarray 이용해 넘파이에서 다차원 배열 쉽게 생성 가능, 다양한 연산 수행 가능.

-**np.array()** 함수

 : ndarray로 변환을 원하는 객체를 인자로 입력 시 ndarray를 반환.

-생성된 ndarray 배열의 **shape** 변수는 ndarray의 크기(행과 열의 수)를 튜플(tuple) 형태로 가지고 있음. -> ndarray 배열의 차원 알 수 있다.

ex1)

```python
array1 = np.array([1,2,3])
print('array1 type:', type(array1))
print('array1 array 형태:', array1.shape)
print('array1 array 차원: ', array1.ndim, '차원')
```

ex2)

```python
array2 = np.array([[1,2,3]])
print('array2 type:', type(array2))
print('array2 array 형태:', array2.shape)
print('array2 array 차원: ', array2.ndim,'차원')
```

##■ ndarray의 데이터 타입

---

-ndarray내의 데이터 타입: **같은 데이터 타입만 가능.**

-다른 데이터 유형이 섞여있는 리스트를 ndarray로 변경 시, **데이터 크기가 더 큰 데이터 타입으로 형 변환.**

-**astype()**: ndarray 내 데이터값의 타입 변경

: 주로 메모리 절약 시 이용.

ex)

```python
array_int = np.array([1,2,3])
print(array_int,array_int.dtype)
```

```python
array_float = array_int.astype('float64')
print(array_float,array_float.dtype)
```

## ■ ndarray를 편리하게 생성하기

---

-특정 크기와 차원을 가진 ndarray를 연속값이나 0 또는 1로 초기화해 쉽게 생성하는 함수들

**: arage(), zeros(), ones()**

**1) arange()**

: range()와 유사한 기능.

ex)

```python
sequence_array = np.arange(10)
print('sequence_array:',sequence_array)
print('sequence_array 형태: ',sequence_array.shape)
```

**2) zeros()**

: 튜플 형태의 shape 값 입력 시 <u>모든 값을 0으로</u> 채운 해당 shape를 가진 ndarray 반환.

ex) zero_array = np.zeros((3,2), dtype = 'int32')

**3) ones()**

: 튜플 형태의 shape 값 입력 시 <u>모든 값을 1로</u> 채운 해당 shape를 가진 ndarray 반환

(dtype 정해주지 않을 시 default: float64)

##■ ndarray의 차원과 크기를 변경하는 reshape()

---

-**reshape()** : ndarray를 특정 차원 및 크기로 변환.

ex)

```python
array1 = np.arange(10)
print('array1:\n',array1)

array2 = array1.reshape(2,5)
print('array2:\n', array2)
```

-인자로 -1 적용하면 <u>알아서 호환되는 shape로 변환해준다.</u>

ex)

```python
array1 = np.arange(10)
print(array1.shape)

array2 = array1.reshape(-1,5)
print('array1.reshape(-1,5)의 형태:',array2.shape)
```

-이런 -1 인자는 reshape(-1,1)과 같은 형태로 자주 사용됨.

-**reshape(-1,1)**은 원본 ndarray가 어떤 형태라도 <u>2차원</u>이고, 여러 개의 row를 가지되 반드시 <u>1개의 column</u>을 가진 ndarray로 변환됨을 보장함.

-**tolist()** : ndarray를 리스트 자료형으로 변환.

##■ 넘파이의 ndarray의 데이터 세트 선택하기 - 인덱싱(Indexing)

---

**1) 특정한 데이터만 추출**

**2) 슬라이싱(Slicing):** 연속된 인덱스상의 ndarray 추출

**3) 팬시 인덱싱(Fancy Indexing):** 일정한 인덱싱 집합을 리스트 또는 ndarray 형태로 지정해 해당 위치에 있는 데이터의 ndarray 변환.

**4) 불린 인덱싱(Boolean Indexing):** 특정 조건에 해당하는지 여부인 True/False 값 인덱싱 집합을 기반을 True에 해당하는 인덱스 위치에 있는 데이터의 ndarray 반환.

-row = axis 0, column = axis 1

-1행 1열 : array1[1,1]

**2) 슬라이싱**

-2차원 슬라이싱

-2차원 ndarray에서 뒤에 오는 인덱스를 없애면 <u>1차원 ndarray로 반환.</u>

ex)

```python
array1d = np.arange(start = 1, stop = 10)
array2d = array1d.reshape(3,3)

print("array2d:\n", array2d)
print('array2d[0]:',array2d[0])
print('array2d[0] shape:', array2d[0].shape)
```

**3) 팬시 인덱싱(Fancy Indexing)**

: 리스트나 ndarray로 인덱스 집합을 지정하면 해당 위치의 인덱스에 해당하는 ndarray를 반환하는 인덱싱 방식

**4) 불린 인덱싱(Boolean Indexing)**

: 조건 필터링과 검색을 동시에 가능한 방식

ex)

```python
array1d = np.arange(start = 1, stop = 10)
array3 = array1d[array1d>5]
print('array1d > 5 Boolean Indexing 결과 값:', array3)
```

-Boolean Indexing이 동작하는 순차적 단계:

Step 1: array1d > 5 와 같이 ndarray의 필터링 조건을 [] 안에 기재

Step 2 : False 값은 무시하고 True 값에 해당하는 인덱스값만 저장

Step 3 : 저장된 인덱스 데이터 세트로 ndarray 조회

##■ 행렬의 정렬 - sort()와 argsort()

---

-넘파이에서 행렬을 정렬하는 대표적 방법: **np.sort(), ndarray.sort()**

-정렬된 행렬의 인덱스 반환하는 방법: **argsort()**

**1) 행렬 정렬**

-np.sort() : 넘파이에서 sort() 호출

-ndarray.sort() : 행렬 자체에서 sort() 호출

-차이: **np.sort()**는 <u>원 행렬 유지한 채</u> 정렬된 행렬 반환, **ndarray.sort()**는 <u>원행렬 자체를 정렬</u>한 형태로 변환, 반환 값은 None

ex)

```python
org_array = np.array([3, 1, 9, 5])
print('원본 행렬:', org_array)

#np.sort()로 정렬
sort_array1 = np.sort(org_array)
print('np.sort() 호출 후 반환된 정렬 행렬:', sort_array1)
print('np.sort() 호출 후 원본 행렬:', org_array)

#ndarray.sort()로 정렬
sort_array2 = org_array.sort()
print('org_array.sort() 호출 후 반환된 행렬:',sort_array2)
print('org_array.sort() 호출 후 원본 행렬:',org_array)
```

-기본값 : 오름차순.

(내림차순 정렬: [::-1])

-행렬이 2차원 이상일 경우, axis 축 값 설정 통해 row 방향, column 방향으로 정렬 수행

**2) 정렬된 행렬의 인덱스를 반환하기**

-np.argsort : 정렬 행렬의 원본 행렬 인덱스를 ndarray 형으로 반환

캡처4.PNG

-기본값 : 오름차순.

(내림차순 정렬: [::-1]

-넘파이는 메타 데이터 가질 수 없음.

예를 들어, 학생별 시험 성적을 데이털 표현하기 위해서 학생의 이름, 시험 성적 <u>각각 ndarray로 가져야함.</u>

즉, John = 78, Mike = 95, Sarah = 84를 ndarray로 활용하고자 한다면

name_array = ['John','Mike','Sarah']와 score_array = [78,95,84]로 2개의 ndarray 만들어야함.

=> argsort()가 넘파이에서 활용도가 높은 이유.

```python
import numpy as np

name_array = np.array(['John','Mike','Sarah','Kate','Samuel'])
score_array = np.array([78,95,84,98,88])

sort_indices_asc = np.argsort(score_array)
print('성적 오름차순 정렬 시 score_array의 인덱스:', sort_indices_asc)
print('성적 오름차순으로 name_array의 이름 출력:', name_array[sort_indices_asc])
```
